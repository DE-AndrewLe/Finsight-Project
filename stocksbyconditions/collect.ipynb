```python
def processing(df):
    df = df.rename(columns={'time': 'Date'})
    df = df.rename(columns={'open': 'Open'})
    df = df.rename(columns={'high': 'High'})
    df = df.rename(columns={'low': 'Low'})
    df = df.rename(columns={'close': 'Close'})
    df = df.rename(columns={'volume': 'Volume'})
    df = df.rename(columns={'ticker': 'Ticker'})

    # Chuyển định dạng cột 'Date' thành tháng/ngày/năm
    df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%m/%d/%Y')

    # Sắp xếp lại các cột
    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]
    cols = ['Open', 'High', 'Low', 'Close']
    for col in cols:
        df[col] = df[col] / 1000
    return df
```

```python
acb1=processing(acb)
bmp1=processing(bmp)
ctg1=processing(ctg)
dcm1=processing(dcm)
gas1=processing(gas)
gmd1=processing(gmd)
hdg1=processing(hdg)
imp1=processing(imp)
acb1
```

```python
print(acb1.dtypes)
acb1.to_csv('ACB1.csv', index=False)
```

```python
def filter_stocks(df):
    df['Date'] = pd.to_datetime(df['Date'])
    # thêm cột
    df['Month'] = df['Date'].dt.month
    df['Quarter'] = df['Date'].dt.quarter
    df['Year'] = df['Date'].dt.year

    # Lọc dữ liệu cho hai quý đầu tiên và không lấy dữ liệu của năm 2020
    df_filtered2quy = df[(df['Quarter'] < 3) & (df['Year'] != 2020)]
    # lấy tháng 3 và 4
    df_filteredmonth34 = df[(df['Month'].isin([3, 4])) & (df['Year'] != 2020)]
    # lấy các tháng 1,2,3,4 từ 2021
    df_filteredmonth1234_2021 = df[(df['Month'].isin([1, 2, 3, 4])) & (df['Year'] >= 2021)]

    # Loại bỏ các hàng có giá trị null
    df_filtered2quy = df_filtered2quy.dropna()
    df_filteredmonth34 = df_filteredmonth34.dropna()
    df_filteredmonth1234_2021 = df_filteredmonth1234_2021.dropna()
    
    # Loại bỏ các hàng trùng lập
    df_filtered2quy = df_filtered2quy.drop_duplicates()
    df_filteredmonth34 = df_filteredmonth34.drop_duplicates()
    df_filteredmonth1234_2021 = df_filteredmonth1234_2021.drop_duplicates()
    
    return df_filtered2quy, df_filteredmonth34, df_filteredmonth1234_2021
```

```python
acb_2quy, acb_month34, acb_month1234_2021 = filter_stocks(acb1)
bmp_2quy, bmp_month34, bmp_month1234_2021 = filter_stocks(bmp1)
ctg_2quy, ctg_month34, ctg_month1234_2021 = filter_stocks(ctg1)
dcm_2quy, dcm_month34, dcm_month1234_2021 = filter_stocks(dcm1)
gas_2quy, gas_month34, gas_month1234_2021 = filter_stocks(gas1)
gmd_2quy, gmd_month34, gmd_month1234_2021 = filter_stocks(gmd1)
hdg_2quy, hdg_month34, hdg_month1234_2021 = filter_stocks(hdg1)
imp_2quy, imp_month34, imp_month1234_2021 = filter_stocks(imp1)
```
