```python
import vnstock
import pandas as pd
import numpy as np
```

```python
acb_up = vnstock.stock_historical_data('ACB','2024-02-02','2024-03-13','1D')
acb_up
```

```python
mbb_up = vnstock.stock_historical_data('MBB','2024-02-02','2024-03-13','1D')
mbb_up
```

```python
ctg_up = vnstock.stock_historical_data('CTG','2024-02-02','2024-03-13','1D')
ctg_up
```

```python
print(acb_up.dtypes)
```

```python
# Đổi tên cột 'Time' thành 'Date'
def processing(df):
    df = df.rename(columns={'time': 'Date'})
    df = df.rename(columns={'open': 'Open'})
    df = df.rename(columns={'high': 'High'})
    df = df.rename(columns={'low': 'Low'})
    df = df.rename(columns={'close': 'Close'})
    df = df.rename(columns={'volume': 'Volume'})
    df = df.rename(columns={'ticker': 'Ticker'})

    # Chuyển định dạng cột 'Date' thành tháng/ngày/năm
    df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%m/%d/%Y')

    # Sắp xếp lại các cột
    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]
    cols = ['Open', 'High', 'Low', 'Close']
    for col in cols:
        df[col] = df[col] / 1000
    return df
```

```python
acb_up=processing(acb_up)
print(acb_up)
acb_up.to_csv('acb_up.csv',index=False)
```

```python
mbb_up=processing(mbb_up)
print(mbb_up)
mbb_up.to_csv('mbb_up.csv',index=False)
```

```python
ctg_up=processing(ctg_up)
print(ctg_up)
ctg_up.to_csv('ctg_up.csv',index=False)
```
